```python
import numpy as np
```

```python
x = np.random.random(1000)
```

```python
print(x)
```

To calculate the mean of this numpy array, sum all the elements in the array and divide them by the total number of elements

```python
mean = sum(x) / len(x)
print(mean)
```

To calculate the mean using the numpy mean function, try the following:


```python
np_mean = np.mean(x)
print(np_mean)
```
